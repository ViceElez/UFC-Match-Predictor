```python
fights["Date"]=pd.to_datetime(fights["Date"])
```

```python
fighters = pd.unique(fights[["RedFighter", "BlueFighter"]].values.ravel())

fighter_map = {name: idx for idx, name in enumerate(fighters)}

fights["RedFighterInt"] = fights["RedFighter"].map(fighter_map)
fights["BlueFighterInt"] = fights["BlueFighter"].map(fighter_map)
```

```python
def get_winner_fighter(row):
    if row['Winner'] == 'Red':
        return fighter_map[row['RedFighter']]
    elif row['Winner'] == 'Blue':
        return fighter_map[row['BlueFighter']]
    else:
        return None 

fights['WinnerFighterInt'] = fights.apply(get_winner_fighter, axis=1)
```

```python
fights["TitleBoutInt"] = fights["TitleBout"].astype(int)
```

```python
fights["WeightClassInt"] = fights["WeightClass"].astype('category').cat.codes
```

```python
fights.head()
```

```python
fights["GenderInt"] = fights["Gender"].astype('category').cat.codes
fights["LocationInt"] = fights["Location"].astype('category').cat.codes
fights["CountryInt"] = fights["Country"].astype('category').cat.codes
fights["RedStanceInt"] = fights["RedStance"].astype('category').cat.codes
fights["BlueStanceInt"] = fights["BlueStance"].astype('category').cat.codes
fights["FinishInt"] = fights["Finish"].astype('category').cat.codes
fights["FinishDetailsInt"] = fights["FinishDetails"].astype('category').cat.codes
```

```python
import numpy as np
def get_better_ranked_fighter(row):
    if row['BetterRank'] == 'Red':
        return fighter_map[row['RedFighter']]
    elif row['BetterRank'] == 'Blue':
        return fighter_map[row['BlueFighter']]
    else:
        return np.nan

fights['BetterRankInt'] = fights.apply(get_better_ranked_fighter, axis=1)
```
