```python
import pandas as pd
```

```python
fights=pd.read_csv("data.csv")
```

```python
pd.set_option('display.max_columns', None)
fights.head()
```

```python
fights["Date"]=pd.to_datetime(fights["Date"])
```

```python
fights["TitleBoutInt"] = fights["TitleBout"].astype(int)
```

```python
fights["WeightClassInt"] = fights["WeightClass"].astype('category').cat.codes
```

```python
fights["GenderInt"] = fights["Gender"].astype('category').cat.codes
fights["LocationInt"] = fights["Location"].astype('category').cat.codes
fights["CountryInt"] = fights["Country"].astype('category').cat.codes
fights["RedStanceInt"] = fights["RedStance"].astype('category').cat.codes
fights["BlueStanceInt"] = fights["BlueStance"].astype('category').cat.codes
fights["BetterRankInt"] = fights["BetterRank"].map({"Red": 1, "Blue": 0}).fillna(-1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100, min_samples_split=10, random_state=1)
```

```python
rf.fit(train[predictors],train["WinnerFighter"])
```

```python
preds=rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test['WinnerFighter'], preds)
```
